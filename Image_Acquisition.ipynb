```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('/content/drive/MyDrive/Orange/orange.png')
```

```python
cv2_imshow(img)
```

```python
#convert to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#display
cv2_imshow(img_gray)
```

```python
#resize image
img_rs = cv2.resize(img, (100, 100))

#display
cv2_imshow(img_rs)
```

```python
#resize image
img_flip = cv2.flip(img, 0)

#display
cv2_imshow(img_flip)
```
